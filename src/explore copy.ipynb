```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Usar la ruta absoluta que mencionaste
file_path = '/workspaces/machine-learning-python-template/data/raw/AB_NYC_2019.csv'

# Verificar si el archivo existe
if os.path.exists(file_path):
    print(f"‚úÖ Archivo encontrado en: {file_path}")
    
    # Cargar datos
    df = pd.read_csv(file_path)
    
    print("=== AN√ÅLISIS INICIAL AIRBNB NYC ===")
    print(f"üìä Dimensiones: {df.shape}")
    print(f"üìù N√∫mero de columnas: {len(df.columns)}")
    
    print("\n=== INFORMACI√ìN DEL DATASET ===")
    df.info()
    
    print("\nüîç Primeras 5 filas:")
    df.head()
    
else:
    print(f"‚ùå Archivo NO encontrado en: {file_path}")
    print("\nüìÅ Buscando archivos en el proyecto...")
    
    # Buscar recursivamente el archivo
    for root, dirs, files in os.walk('/workspaces/machine-learning-python-template'):
        for file in files:
            if 'AB_NYC' in file or 'airbnb' in file.lower():
                print(f"üìÑ Encontrado: {os.path.join(root, file)}")
```

```python
print("=== AN√ÅLISIS DESCRIPTIVO ===")
print(df.describe())

print("\n=== VALORES NULOS ===")
print(df.isnull().sum().sort_values(ascending=False))

print("\n=== TIPOS DE HABITACI√ìN ===")
print(df['room_type'].value_counts())

print("\n=== DISTRIBUCI√ìN POR BARRIOS ===")
print(df['neighbourhood_group'].value_counts())
```

```python
# Configurar el estilo de los gr√°ficos
plt.style.use('default')
sns.set_palette("husl")

# 1. Distribuci√≥n de precios
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
sns.histplot(df['price'], bins=50)
plt.title('Distribuci√≥n de Precios')
plt.xlabel('Precio ($)')
plt.xlim(0, 1000)  # Excluir outliers extremos

# 2. Tipos de habitaci√≥n
plt.subplot(2, 3, 2)
df['room_type'].value_counts().plot(kind='bar')
plt.title('Tipos de Habitaci√≥n')
plt.xticks(rotation=45)

# 3. Distribuci√≥n por barrios
plt.subplot(2, 3, 3)
df['neighbourhood_group'].value_counts().plot(kind='bar')
plt.title('Propiedades por Barrio')
plt.xticks(rotation=45)

# 4. Precio promedio por barrio
plt.subplot(2, 3, 4)
sns.barplot(data=df, x='neighbourhood_group', y='price')
plt.title('Precio Promedio por Barrio')
plt.xticks(rotation=45)

# 5. Precio por tipo de habitaci√≥n
plt.subplot(2, 3, 5)
sns.boxplot(data=df, x='room_type', y='price')
plt.title('Precio por Tipo de Habitaci√≥n')
plt.xticks(rotation=45)
plt.ylim(0, 500)

# 6. Mapa de calor de correlaciones
plt.subplot(2, 3, 6)
numeric_cols = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365']
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlaciones entre Variables')

plt.tight_layout()
plt.show()
```
